Contar numero de terminos/palabras en el corpues


```python
       
import math

def Conteo():
    palabras_Total = 0  # Inicializamos el contador de palabras
    try: 
        with open('01_corpus_turismo.txt', 'r') as file:
            for line in file:
                if not line.strip():
                    continue
                palabras = dividir(line)  # Usamos la misma función dividir que limpia y divide
                palabras_Total += len(palabras)  # Contamos las palabras en la línea
            return palabras_Total
    except FileNotFoundError:
        print("El archivo no se encontró.")

def construir_matriz_termino_documento(archivo):
    matriz_termino_documento = {}
    documentos = []  # Lista para almacenar los documentos
    try:
        with open(archivo, 'r', encoding='utf-8') as file:
            index = 1
            for line in file:
                documentos.append(index)  # Agregamos el índice del documento a la lista
                if not line.strip():
                    continue
                palabras = dividir(line) 
                for palabra in palabras:
                    if palabra not in matriz_termino_documento:
                        matriz_termino_documento[palabra] = {}
                    if index not in matriz_termino_documento[palabra]:
                        matriz_termino_documento[palabra][index] = 1
                    else:
                        matriz_termino_documento[palabra][index] += 1
                index += 1
                
            for palabra in matriz_termino_documento:
                for doc in documentos:
                    if doc not in matriz_termino_documento[palabra]:
                        matriz_termino_documento[palabra][doc] = 0
    except FileNotFoundError:
        print(f"Error: El archivo '{archivo}' no se encontró.")
    
    return matriz_termino_documento, documentos

def palabras_en_documento(matriz_termino_documento):
    total_palabras_en_documento = {}
    for palabra in matriz_termino_documento:
        for doc, frecuencia in matriz_termino_documento[palabra].items():
            if doc not in total_palabras_en_documento:
                total_palabras_en_documento[doc] = 0
            total_palabras_en_documento[doc] += frecuencia
    return total_palabras_en_documento

def calcular_tf_idf(matriz, total_palabras_en_documento, total_documentos):
    tf_total = {}   
    idf_total = {} 

    for palabra in matriz:
        tf_total[palabra] = {}
        documentos_con_palabra = sum(1 for doc in matriz[palabra] if matriz[palabra][doc] > 0)
        
        # Calculamos el IDF solo una vez por palabra
        idf = math.log(total_documentos / documentos_con_palabra)
        idf_total[palabra] = idf

        for doc in matriz[palabra]:
            tf = matriz[palabra][doc] / total_palabras_en_documento[doc]
            tf_total[palabra][doc] = tf

    return tf_total, idf_total

#Esta funcion se encarga de limpiar texto y dividrlo en palabras si hay algun caracter especial con tilde lo reemplazara con el mismo caracter 
#sin tilde si es otro lo tomara como espacio
def dividir(texto):
    texto = texto.lower()
    texto_limpio = ""
    for caracter in texto:
        if caracter.isalnum() or caracter in "áéíóúüñ": #isalnum es booleano, true cuando el caracter es alfanumérico y flase cuando no lo es
            texto_limpio += caracter
        else:
            texto_limpio += " "
    palabras = texto_limpio.split()
    return palabras 

#IMPRESIONES

def imprimir_matriz(matriz, documentos):
    print(f"{'Término':<15}", end="")
    for doc_id in documentos:
        print(f"Doc {doc_id:<8}", end="")
    print()
    
    for termino in matriz:
        print(f"{termino:<15}", end="")
        for doc_id in documentos:
            print(f"{matriz[termino].get(doc_id, 0):<10}", end="")
        print()

def imprimir_tf(tf, documentos):
    print("\nTF (Term Frequency):")
    print(f"{'Término':<15}", end="")
    for doc_id in documentos:
        print(f"Doc {doc_id:<8}", end="")
    print()
    
    for termino in tf:
        print(f"{termino:<15}", end="")
        for doc_id in documentos:
            valor = tf[termino].get(doc_id, 0)
            print(f"{valor:<10.4f}", end="")
        print()

def imprimir_idf(idf):
    print("\nIDF (Inverse Document Frequency):")
    print(f"{'Término':<15}IDF")
    for termino in idf:
        print(f"{termino:<15}{idf[termino]:.4f}")


#Prueba de la función Conteo
palabras_Total = Conteo()
matriz_termino_documento, documentos = construir_matriz_termino_documento('01_corpus_turismo.txt')
total_palabra_en_documento = palabras_en_documento(matriz_termino_documento)
total_documentos = len(documentos)  # Total de documentos procesados

tf_total, idf_total = calcular_tf_idf(matriz_termino_documento, total_palabra_en_documento, total_documentos)


#print(f"TF total: {tf_total}")
#print(f"IDF total: {idf_total}")
print(f"Total de documentos: {total_documentos}")
#print(f"Conteo de palabras en el documento: {total_palabra_en_documento}")
#print(f"construir_matriz_termino_documento: {matriz_termino_documento}")
print(f"Cantidad de palabras totales en el documento: {palabras_Total}")

imprimir_matriz(matriz_termino_documento, documentos)
imprimir_tf(tf_total, documentos)
imprimir_idf(idf_total)

```
